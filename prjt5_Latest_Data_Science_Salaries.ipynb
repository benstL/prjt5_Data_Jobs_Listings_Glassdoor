```python
import pandas as pd
```

```python


# Load the datasets
df1 = pd.read_csv("Latest_Data_Science_Salaries.csv")
df2 = pd.read_csv("v2_Latest_Data_Science_Salaries.csv")
df3 = pd.read_csv("v3_Latest_Data_Science_Salaries.csv")

# Concatenate the datasets vertically
merged_df = pd.concat([df1, df2, df3], ignore_index=True)

# Normalize column names by converting to lowercase and removing spaces
merged_df.columns = merged_df.columns.str.lower().str.replace(" ", "_")

# Calculate the frequency of each job_title
job_title_counts = merged_df['job_title'].value_counts()

# Create a new column to store the standardized salary values
merged_df['standardized_salary'] = 0.0

# Standardize salary within each job_title group
for job_title, count in job_title_counts.items():
    mask = merged_df['job_title'] == job_title
    merged_df.loc[mask, 'standardized_salary'] = (merged_df.loc[mask, 'salary'] - merged_df.loc[mask, 'salary'].mean()) / merged_df.loc[mask, 'salary'].std()

# Save the standardized dataframe to a new CSV file
merged_df.to_csv('Standardized_Data.csv', index=False)

# Display the standardized dataframe
display(merged_df)
```
